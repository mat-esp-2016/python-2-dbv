```python
import numpy as np
import matplotlib.pyplot as plt
import glob as gb
```

Primeiro, importamos as bibliotecas numpy, matplotlib.pyplot e glob, apelidando-as como np, plt e gb respectivamente.

```python
%matplotlib inline
```

Chamamos a função inline do matplotlib para que os gráficos sejam exibidos na célula segunte ao código.

```python
arquivos = gb.glob("dados/*.txt")
for arquivo in arquivos:
    if arquivo != 'dados/brazil-TAVG-Trend.txt':
        dados = np.loadtxt(arquivo, comments="%")
        anomalia_temperatura_anual = dados[:,4]
        anos = dados[:,0]
        meses = dados[:,1]
        tempo = ( anos + ( meses / 12))
        plt.ylabel("Mean Temperature(°C)")
        plt.plot(tempo,anomalia_temperatura_anual,color='black',label='12-month moving average')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.show()
        arquivo_png = arquivo[0:-3] + "png"
        plt.savefig(arquivo_png)
```

Algorítimo:

<p>1 - Carregamos uma lista chamada arquivos contendo o nome de todos os arquivos com o sufixo ".txt" no diretório dados.</p>
<p>2 - Criamos um laço de repetição for para trabalhar com cada nome de arquivo dentro da lista arquivos. Caso o nome do arquivo seja diferente de "dados/brazil-TAVG-Trend.txt", ele gera o gráfico.</p>
<p>3 - Para cada arquivo, carregamos os dados deste arquivo na lista dados.</p>
<p>4 - Criamos uma lista chamada anomalia_temperatura_anual contendo todas as anomalias de temperatura de todo os anos daquele arquivo. O mesmo é feito para  anos e meses.</p>
Para que o gráfico seja gerado corretamente, criamos uma outra lista chamada tempo contendo a soma dos anos com a fração do ano(meses), transformando anos e mêses em apenas anos:</p>
$$tempo = anos + \frac{meses}{12}$$
<p>5 - Criamos o gráfico contendo o a anomalia de temperatura de todos os anos em função do tempo.</p>
<p>6 - Salvamos cada gráfico com a função savefig da biblioteca matlibplot.pyplot, modificando o nome do arquivo para conter o sufixo "png" no lugar de "txt".</p>

```python
arquivos = gb.glob("dados/*.txt")
for arquivo in arquivos:
    if arquivo != 'dados/brazil-TAVG-Trend.txt':
        dados = np.loadtxt(arquivo, comments="%")
        anomalia_temperatura_anual = dados[:,4]
        anos = dados[:,0]
        meses = dados[:,1]
        tempo = ( anos + ( meses / 12))
        anomalia_temperatura_10anos = np.nan_to_num(dados[:,8])
        incerteza_anomalia_temperatura_10anos = np.nan_to_num(dados[:,9])
        y1 = anomalia_temperatura_10anos + incerteza_anomalia_temperatura_10anos
        y2 = anomalia_temperatura_10anos - incerteza_anomalia_temperatura_10anos
        plt.ylabel("Mean Temperature(°C)")
        plt.fill_between(tempo,y1,y2,color='gray',)
        plt.plot(tempo,anomalia_temperatura_10anos,color='red',label='10-year moving average with 95% uncertainty range')
        plt.plot(tempo,anomalia_temperatura_anual,color='black',label='12-month moving average')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.show()
```
