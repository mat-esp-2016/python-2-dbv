```python
import numpy as np
import matplotlib.pyplot as plt
import glob as gb
```

```python
%matplotlib inline
```

```python
arquivos = gb.glob("dados/*.txt")
for arquivo in arquivos:
    if arquivo != 'dados/brazil-TAVG-Trend.txt':
        dados = np.loadtxt(arquivo, comments="%")
        anomalia_temperatura_anual = dados[:,4]
        anos = dados[:,0]
        meses = dados[:,1]
        tempo = ( anos + ( meses / 12))
        plt.ylabel("Mean Temperature(C)")
        plt.plot(tempo,anomalia_temperatura_anual,color='black',label='12-month moving average')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.show()
        arquivo_png = arquivo[0:-3] + "png"
        plt.savefig(arquivo_png)
```

```python
arquivos = gb.glob("dados/*.txt")
for arquivo in arquivos:
    if arquivo != 'dados/brazil-TAVG-Trend.txt':
        dados = np.loadtxt(arquivo, comments="%")
        anomalia_temperatura_anual = dados[:,4]
        anos = dados[:,0]
        meses = dados[:,1]
        tempo = ( anos + ( meses / 12))
        anomalia_temperatura_10anos = dados[:,8]
        plt.ylabel("Mean Temperature(C)")
        plt.fill_between(tempo,anomalia_temperatura_10anos,color='gray',)
        plt.plot(tempo,anomalia_temperatura_10anos,color='red',label='10-year moving average')
        plt.plot(tempo,anomalia_temperatura_anual,color='black',label='12-month moving average')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.show()
```
